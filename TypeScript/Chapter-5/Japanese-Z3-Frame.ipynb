# Japanese IQ Test

The following puzzle is part of a [recruitment test](https://flashmuseum.org/iq-%E9%81%8E%E6%B2%B3%E9%81%8A%E6%88%B2/) in Japan.

---

A policeman, a convict, a father and his two sons Anton and Bruno, and a mother with her two daughters Cindy and Doris have to cross a river.  On the boat there is only room for two passengers.

During the crossing, the following conditions have to be observed:
* The father is not allowed to be on a shore with one of the daughters 
  if the mother is on the other shore. 
* The mother is not allowed to be on a shore with one of the sons 
  if the father is on the other shore.
* If the criminal is not alone, then the policeman must watch him.

  However the convict can be alone on a shore, as his shackles prevent him from running away. 
* Only the father, the mother, and the policeman are able to steer the boat. 

We are going to formulate this problem as a *symbolic transition problem*.
Then, we can solve the problem using our constraint solver.

We will use the following variables to encode the problem:
* `P` equals `1` if the policeman is on the western shore, while `PX` represents the variable `P'`.
* `V` equals `1` if the villain is on the western shore, while `VX` represents the variable `V'`
* `M` equals `1` if the mother is on the western shore, while `MX` represents the variable `M'`
* `F` equals `1` if the father is on the western shore, while `FX` represents the variable `F'`
* `B` is the number of boys on the western shore, while `BX` represents the variable `B'`
* `G` is the number of girls on the western shore, while `GX` represents the variable `G'`.

```typescript
async function japanese_CSP(n) {
    const solver = new Z3.Solver();
    
    const Ps = range(n).map(i => Z3.Int.const(`P${i}`));
    const Vs = range(n).map(i => Z3.Int.const(`V${i}`));
    const Fs = range(n).map(i => Z3.Int.const(`F${i}`));
    const Ms = range(n).map(i => Z3.Int.const(`M${i}`));
    const Bs = range(n).map(i => Z3.Int.const(`B${i}`));
    const Gs = range(n).map(i => Z3.Int.const(`G${i}`));

    solver.add(...start(Ps[0], Vs[0], Ms[0], Fs[0], Bs[0], Gs[0]));
    solver.add(...goal (Ps[n], Vs[n], Ms[n], Fs[n], Bs[n], Gs[n]));

    range(n-1).map(i => {
        solver.add(
            "your code here"
            Ps[i].ge(0), Vs[i].ge(0), Fs[i].ge(0), Ms[i].ge(0), Bs[i].ge(0), Gs[i].ge(0),
            Ps[i].le(1), Vs[i].le(1), Fs[i].le(1), Ms[i].le(1), Bs[i].le(2), Gs[i].le(2)
        );
    });

    const result = await solver.check();
    if (result === 'sat') {
        const model = solver.model();   
        const extract = (arr, prefix) => {
            const res = {};
            for (let i = 0; i <= n; i++) {
                res[`${prefix}${i}`] = Number((model.eval(arr[i]) as any).value());
            }
            return res;
        };     
        return {
            ...extract(Ps, 'P'), ...extract(Vs, 'V'), ...extract(Ms, 'M'), 
            ...extract(Fs, 'F'), ...extract(Bs, 'B'), ...extract(Gs, 'G')
        };
    }
    return null;
}
```

```typescript
async function find_solution() {
    let n = 1;
    while (true) {
        console.log(`Checking n = ${n}...`);
        const Solution = await japanese_CSP(n);
        if (Solution !== null) {
            return { n, Solution };
        }
        n += 2;
    }
}
```

```typescript
import * as tslab from "tslab";

function show_animated_solution(solution: Record<string, number>, n: number) {
    if (!solution) {
        console.error("No solution provided.");
        return;
    }

    // Prepare the state data for the frontend JavaScript
    const states = [];
    for (let i = 0; i <= n; i++) {
        states.push({ 
            P: solution[`P${i}`], 
            V: solution[`V${i}`], 
            M: solution[`M${i}`], 
            F: solution[`F${i}`], 
            B: solution[`B${i}`], 
            G: solution[`G${i}`] 
        });
    }

    // Generate a unique ID suffix to avoid element collisions if run multiple times
    const uid = Math.random().toString(36).substring(2, 9);

    const htmlContent = `
    <div style="font-family: sans-serif; max-width: 800px; margin: 0 auto; text-align: center;">
        <div style="margin-bottom: 15px;">
            <button id="btn-reset-${uid}" style="padding: 8px 16px; cursor: pointer;">Reset</button>
            <button id="btn-prev-${uid}" style="padding: 8px 16px; cursor: pointer;">Previous Step</button>
            <button id="btn-next-${uid}" style="padding: 8px 16px; cursor: pointer; font-weight: bold;">Next Step</button>
            <span id="label-step-${uid}" style="margin-left: 15px; font-weight: bold; font-size: 16px;">Step 0 of ${n}</span>
        </div>
        
        <svg width="800" height="250" style="background-color: #e0f7fa; border: 1px solid #b2ebf2; border-radius: 8px;">
            <rect x="0" y="0" width="250" height="250" fill="#a5d6a7" />
            <rect x="550" y="0" width="250" height="250" fill="#a5d6a7" />
            
            <text x="125" y="30" font-size="14" font-weight="bold" fill="#2e7d32" text-anchor="middle">Western Shore</text>
            <text x="675" y="30" font-size="14" font-weight="bold" fill="#2e7d32" text-anchor="middle">Eastern Shore</text>

            <text id="west-people-${uid}" x="125" y="125" font-size="28" text-anchor="middle"></text>
            <text id="east-people-${uid}" x="675" y="125" font-size="28" text-anchor="middle"></text>

            <g id="boat-${uid}" style="transition: transform 1.5s ease-in-out;">
                <rect x="260" y="105" width="120" height="40" rx="8" fill="#8d6e63" stroke="#5d4037" stroke-width="2"/>
                <text id="boat-p-${uid}" x="320" y="133" font-size="22" text-anchor="middle"></text>
            </g>
        </svg>
    </div>

    <script>
    (function() {
        const states = ${JSON.stringify(states)};
        let currentStep = 0;
        let isAnimating = false;
        
        const boat = document.getElementById('boat-${uid}');
        const westPeople = document.getElementById('west-people-${uid}');
        const eastPeople = document.getElementById('east-people-${uid}');
        const boatP = document.getElementById('boat-p-${uid}');
        const stepLabel = document.getElementById('label-step-${uid}');
        
        // Web Audio API context for the motor sound
        let audioCtx;
        function playMotorSound() {
            if (!audioCtx) audioCtx = new (window.AudioContext || window.webkitAudioContext)();
            if (audioCtx.state === 'suspended') audioCtx.resume();
            
            const osc = audioCtx.createOscillator();
            const filter = audioCtx.createBiquadFilter();
            const gain = audioCtx.createGain();
            
            osc.type = 'sawtooth';
            osc.frequency.value = 12; 
            
            filter.type = 'lowpass';
            filter.frequency.value = 150;
            gain.gain.setValueAtTime(0, audioCtx.currentTime);
            gain.gain.linearRampToValueAtTime(0.6, audioCtx.currentTime + 0.2); 
            
            osc.connect(filter);
            filter.connect(gain);
            gain.connect(audioCtx.destination);
            osc.start();
            
            return { osc, gain };
        }
        
        function stopMotorSound(sound) {
            if (sound) {
                sound.gain.gain.linearRampToValueAtTime(0, audioCtx.currentTime + 0.4); 
                setTimeout(() => sound.osc.stop(), 400);
            }
        }

        function makeStr(P, V, M, F, B, G) {
            return '👮'.repeat(P) + '🦹'.repeat(V) + '👩'.repeat(M) + '👨🏿'.repeat(F) + '👦'.repeat(B) + '👧🏾'.repeat(G);
        }

        function renderStaticState(step) {
            const s = states[step];
            westPeople.textContent = makeStr(s.P, s.V, s.M, s.F, s.B, s.G);
            eastPeople.textContent = makeStr(1 - s.P, 1 - s.V, 1 - s.M, 1 - s.F, 2 - s.B, 2 - s.G);
            boatP.textContent = '';
            
            // step % 2 === 0 means West (0px). step % 2 !== 0 means East (170px to touch the other shore).
            boat.style.transform = step % 2 === 0 ? 'translateX(0px)' : 'translateX(170px)';
            stepLabel.textContent = 'Step ' + step + ' of ${n}';
        }

        async function animateTransition(fromStep, toStep) {
            if (isAnimating) return;
            isAnimating = true;
            
            const s0 = states[fromStep];
            const s1 = states[toStep];
            
            const pP = Math.abs(s0.P - s1.P);
            const pV = Math.abs(s0.V - s1.V);
            const pM = Math.abs(s0.M - s1.M);
            const pF = Math.abs(s0.F - s1.F);
            const pB = Math.abs(s0.B - s1.B);
            const pG = Math.abs(s0.G - s1.G);
            
            if (fromStep % 2 === 0) { 
                // West to East
                westPeople.textContent = makeStr(s0.P - pP, s0.V - pV, s0.M - pM, s0.F - pF, s0.B - pB, s0.G - pG);
            } else { 
                // East to West
                eastPeople.textContent = makeStr((1 - s0.P) - pP, (1 - s0.V) - pV, (1 - s0.M) - pM, (1 - s0.F) - pF, (2 - s0.B) - pB, (2 - s0.G) - pG);
            }
            
            boatP.textContent = makeStr(pP, pV, pM, pF, pB, pG);
            
            const sound = playMotorSound();
            boat.style.transform = toStep % 2 === 0 ? 'translateX(0px)' : 'translateX(170px)';
            
            await new Promise(r => setTimeout(r, 1500));
            
            stopMotorSound(sound);
            currentStep = toStep;
            renderStaticState(currentStep);
            isAnimating = false;
        }

        document.getElementById('btn-next-${uid}').addEventListener('click', () => {
            if (currentStep < states.length - 1) animateTransition(currentStep, currentStep + 1);
        });

        document.getElementById('btn-prev-${uid}').addEventListener('click', () => {
            if (currentStep > 0) animateTransition(currentStep, currentStep - 1);
        });

        document.getElementById('btn-reset-${uid}').addEventListener('click', () => {
            if (!isAnimating) {
                currentStep = 0;
                renderStaticState(0);
            }
        });

        renderStaticState(0);
    })();
    </script>
    `;

    tslab.display.html(htmlContent);
}
```
