```typescript
import { RecursiveSet as Set, Value } from 'recursive-set';
```

```typescript
function set<T extends Value>(...elements: T[]): Set<T> {
    return new Set(...elements);
}
```

```typescript
function range(start: number, stop: number): Set<number> {
  return set(...Array.from({ length: stop - start + 1 }, 
                           (_, index) => start + index));
}
```

Given a natural number `n`, the function `divisors(n)` computes the set of all 
natural numbers that divide `n`:
$$ \texttt{divisors}(n) = \{ i \in \mathbb{N} \mid n \,\texttt{%}\, i = 0 \} $$

```typescript
function divisors(n: number): Set<number> {
    return range(1, n-1).filter(i => n % i == 0);
}
```

```typescript
divisors(24);
```

A natural number $n$ is perfect if $n$ is equal to the sum of its proper divisors:
$$ \sum \bigl\{t \mid t \in \{1,\cdots, n-1\} \wedge n \;\%\; t = 0 \bigr\}  = n $$
For example, $6$ is perfect, because $6 = 1 + 2 + 3$, but $8$ is not perfect, because 
$1 + 2 + 4 = 7 \not= 8$.

```typescript
function perfect(n: number): boolean {
    return divisors(n).reduce((s, i) => s + i, 0) == n;
}
```

```typescript
perfect(6)
```

```typescript
perfect(8)
```

```typescript
range(1, 10_000).filter(n => perfect(n));
```

```typescript
function findOddPerfect(): number {
    let n = 1;
    while (true) {
        if (perfect(n)) {
            return n;
        }
        if (n % 1000 == 1) {
            console.log(n);
        }
        n += 2;
    }
}
```
